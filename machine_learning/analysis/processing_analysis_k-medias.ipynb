```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = data.loc[0:,['sale_day','customer_score','losses_day','Organization','Responsibility','Conduct','Resource_management','Conflict_management','Contribution']].values
```

```python
dataset
```

```python
plt.scatter(dataset[:][:,0], dataset[:][:,1],dataset[:][:,2],dataset[:][:,3],)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
modelo = KMeans(n_clusters=2)
```
